##### Now,proceeding for data preprocessing

```python
date_time=d.datetime.now()
date_time.year-data['Year']
```

It will fetch the system year or current year and subtract with the dataset year to calculate the age of the car.

```python
data['Age']=date_time.year-data['Year']
data.head()
```

Now we don't need this Year column in dataset,so that we will remove that column 

```python
data.drop('Year',axis=1,inplace=True)
data.head()
```

##### Now,column removed , so  we will remove the outliers from the dataset 

```python
s.boxplot(data['Selling_Price'])
```

##### As we can see that,two data points are far from the other data points.we will consider those two pints as outliers.Now proceed to see which data will contain these two dataset.

#### Now,by sorting the data from bottom will give those outliers..

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data[(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

##### we can see  these outliers.Now we will remove these two data from dataset

```python
data=data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
data.shape
```

#### Now,those data has been removed and reflected on its shape

##### We have to perform encoding for the categorical data

```python
data.head(1)
```

##### Here,(Fuel_Type	Seller_Type	 Transmission) are categorial data.We know that machine learning only numerical values  so , we have to convert these categorial data into numerical

```python
data['Fuel_Type'].unique()
```

#### Using map ,we will convert its nature.

```python
data['Fuel_Type']=data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
data['Fuel_Type'].unique()
```

##### Now,we can see that the categorial data for fuel_Type has been changed ,so Similaryly we will do the same with rest of the categorical data.

```python
data['Seller_Type'].unique()
```

```python
data['Transmission'].unique()
```

```python
data['Seller_Type']=data['Seller_Type'].map({'Dealer':0,'Individual':1})
```

```python
data['Transmission']=data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
data['Seller_Type'].unique()
```

```python
data['Transmission'].unique()
```

##### Now,we will store Feature in X and Target in vector y where selling_price is our target

```python
X=data.drop(['Car_Name','Selling_Price'],axis=1)
y=data['Selling_Price']
```

```python
X
```

```python
y
```

##### Now,we will do the most important operation that is spliting the data into training data and testing data ,to improve the performance

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.24,random_state=49)
```

##### we will train our data and test it then we will get our prediction by X_test and will get our model by y_test

#### Now,we will import the models(Algos)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble._forest import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

##### Now,will start model training

```python
l=LinearRegression()
l.fit(X_train,y_train)
```

```python
r=RandomForestRegressor()
r.fit(X_train,y_train)
```

```python
d=DecisionTreeRegressor()
d.fit(X_train,y_train)
```

```python
x=XGBRegressor()
x.fit(X_train,y_train)
```

##### So,we have performed our training on data ,now it's time to test data and get the prediction

```python
y_predict1=l.predict(X_test)
y_predict2=r.predict(X_test)
y_predict3=d.predict(X_test)
y_predict4=x.predict(X_test)
```

##### Now,evaluating the performance of the dataset by algos, (r_sqaure score) 

```python
from sklearn import metrics
```

```python
score_1=metrics.r2_score(y_test,y_predict1)
score_2=metrics.r2_score(y_test,y_predict2)
score_3=metrics.r2_score(y_test,y_predict3)
score_4=metrics.r2_score(y_test,y_predict4)
```

```python
Data=pd.DataFrame({'Models':['LinearRegression','RandomForestRegressor','DecisionTreeRegressor','XGBRegressor'],
"R2_Score":[score_1*100,score_2*100,score_3*100,score_4*100]
})
Data
```

##### By looking the score , we can easy find that RandomForestRegressor method or model is suitable and accurate,so we will proceed with this model.

```python
final_model=RandomForestRegressor()
final_model.fit(X_train,y_train)
```

##### Now we will save this model using joblib for further executions

```python
import joblib
```

```python
joblib.dump(final_model,'Prediction_Model')
```

```python
model=joblib.load('Prediction_Model')
```
